###`Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
# write your code here
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
temp_list = []
for a,b in test_list[:3]:
  if a not in temp_list:
    temp_list.append(a)
  if b not in temp_list:
    temp_list.append(b)
  test_list.pop(0)
t=tuple(temp_list)
test_list.insert(0,0)
test_list[0]=t
print(test_list)
````

###`Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# write your code here
original_tuple=(1, 5, 7, 8, 10)
temp_list = []
for i in range(len(original_tuple)):
  if i == 0:
    temp_list.append(original_tuple[i]*original_tuple[i+1])
  elif i == len(original_tuple)-1:
    temp_list.append(original_tuple[i-1]*original_tuple[i])
  else:
    temp_list.append(original_tuple[i-1]*original_tuple[i]+original_tuple[i]*original_tuple[i+1])
print(tuple(temp_list))
````

###`Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
from posixpath import sep
# write your code here
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]
list_count=0
tuple_count=0
set_count=0
for item in list1:
  if isinstance(item,tuple):
    tuple_count +=1
  if isinstance(item,list):
    list_count +=1
  if isinstance(item,set):
    set_count +=1

print('List-',list_count)
print('Set-',set_count)
print('Tuples-',tuple_count)
````

### `Q5`: Sort Dictionary key and values List.

**Example 1:**

Input:

```bash
{'c': [3], 'b': [12, 10], 'a': [19, 4]}
```

Output:

```bash
{'a': [4, 19], 'b': [10, 12], 'c': [3]}
```

**Example 2:**

Input:

```bash
{'c': [10, 34, 3]}
```

Output:

```bash
{'c': [3, 10, 34]}
```

````python
# write your code here
#Note: By Default the dictionary will be soreted based on keys. To do the sorting based on values use lamda function in sorted(iterable,key=lambda x: x[1])
input_dict = {'c': [3], 'b': [100, 10], 'a': [19, 4]}
print("Dictionary sorting based on value: ",dict(sorted(input_dict.items(),key=lambda x:x[1])))
x=dict(sorted(input_dict.items()))
print("By defaul sorting based on Keys:",x)
for key,value in x.items():
  x[key].sort()
print("Desired result a/q: ",x)
````
